```python
import yfinance as yf
import pmdarima as pm
import pandas
from datetime import datetime

data = yf.download('%5ENSEI', start='2014-01-01', end=datetime.now().strftime('%Y-%m-%d'))

  
```

```python
data[f'OPENHIGH'] = (data['Open'] - data['High']).abs()
data[f'OPENLOW'] = (data['Open'] - data['Low']).abs()
data[f'HIGHLOW'] = (data['High'] - data['Low']).abs()
```

```python
data['P_HL'] = (data['HIGHLOW'] / data['Open'])*100
data['P_OH'] = (data['OPENHIGH'] / data['Open'])*100
data['P_OL'] = (data['OPENLOW'] / data['Open'])*100
```

```python
data[['P_HL','P_OH','P_OL']]
```

```python
# result = pm.auto_arima(data.HIGH-LOW,d=0,start_p=0,start_q=0,max_p=6,max_q=6,seasonal=True,m=12,D=4,start_P=1,start_Q=1,
#                            max_P=6,max_Q=6,information_criterion='aic',trace=True,stepwise=True)

p_oh = pm.auto_arima(data.P_OH, trace=True)
p_ol = pm.auto_arima(data.P_OL, trace=True)
```

```python
p_oh, p_ol
```

```python
import statsmodels.api as sm
```

```python
train =  data.iloc[:2000]
test = data.iloc[2000:]
```

```python
train.shape, test.shape
```

```python
p_oh_model = sm.tsa.arima.ARIMA(train['P_OH'],order=(1, 0, 1))
p_ol_model = sm.tsa.arima.ARIMA(train['P_OL'],order=(4, 1, 1))


p_oh_model = p_oh_model.fit()
p_ol_model = p_ol_model.fit()
```

```python
pandas.set_option('display.max_columns', None)  # or 1000
pandas.set_option('display.max_rows', None)  # or 1000
pandas.set_option('display.max_colwidth', None)  # or 199
```

```python

start = len(train)
end = len(train)+len(test)-1


pred_oh =  p_oh_model.predict(start=start,end=end,type='levels')
pred_ol =  p_ol_model.predict(start=start,end=end,type='levels')


pred_oh.index = data.index[start:end+1]
pred_ol.index = data.index[start:end+1]

pred_df = pandas.concat([data,pred_oh,pred_ol],axis=1)
pred_df = pred_df.iloc[start:]

# pred_df['Date','OPEN','HIGH','LOW','','',
```
